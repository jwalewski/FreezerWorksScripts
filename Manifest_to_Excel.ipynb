```python
#Part one: Instantiate Filepaths
input_filepath=""
output_serum_filepath=""
output_pbmc_filepath=""

#Part two: Begin Data Processing
old_dataframe = pandas.read_csv("", header=1) #sets header to second row
old_dataframe
serum_array=[]#holds each line as a string
pbmc_array=[]#they have commas in between and newlines at the end
total_serum_count=0 #Every 10 samples we start a new row in the freezer
total_pbmc_count=0
for row in range(len(old_dataframe)):
    date = old_dataframe.iloc[row,0]
    patient_id=old_dataframe.iloc[row,1]
    current_pbmc_aliquot_number=old_dataframe.iloc[row,2]
    pbmc_quantity=old_dataframe.iloc[row,3]
    current_serum_number=old_dataframe.iloc[row,4]
    # print(current_serum_number)
    # print(type(current_serum_number))
    if not(numpy.isnan(current_serum_number)):
        for i in range(int(current_serum_number)):
            total_serum_count+=1
            #print(f"current_serum_number is: {current_serum_number} and we are printing this the {i}th time")
            #print(f"The string to be added will be: {patient_id} {date} .5mL,")
            serum_array.append(str(patient_id)+" "+str(date)+" .5mL,")
            if total_serum_count%10 ==0:
                serum_array.append("\n")
    if not(numpy.isnan(current_pbmc_aliquot_number)):                                                                                                                                                                                                                                                                                                            
        for j in range(int(current_pbmc_aliquot_number)):
            total_pbmc_count+=1
            #print(f"current_pbmc_aliquot_number is: {current_pbmc_aliquot_number}  and we are printing this the {j}th time")
            #print(f"The string to be added will be: {patient_id} {date} {pbmc_quantity},")
            pbmc_array.append(str(patient_id)+" "+str(date)+" "+str(pbmc_quantity)+",")
            if total_pbmc_count%10 ==0:
                pbmc_array.append("\n")

with open(output_serum_filepath, "w") as output_serum_file:
    for line in serum_array:
        output_serum_file.write(line)
with open(output_pbmc_filepath, "w") as output_pbmc_file:
    for line in pbmc_array:
        output_pbmc_file.write(line)
```
